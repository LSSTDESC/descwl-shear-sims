```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
sns.set()
```

```python
from descwl_shear_sims.simple_sim import Sim
```

## Make A Sim w/ a Crazy WCS

```python
data = Sim(
    rng=1000, 
    epochs_per_band=3,
    wcs_kws=dict(
        position_angle_range=(0, 360),
        scale_frac_std=0.1,
        dither_range=(-0.5, 0.5),
        shear_std=0.1)).gen_sim()
```

```python
fig, axs  = plt.subplots(nrows=3, ncols=3, figsize=(12, 12))

for band_ind, band in enumerate(data):
    for epoch_ind, obs in enumerate(data[band]):
        ax = axs[band_ind, epoch_ind]
        ax.imshow(obs.image.array)
        ax.grid(False)
```

## Make a Sim w/ a Grid of Objects

```python
data = Sim(
    rng=1000, 
    epochs_per_band=3,
    grid_gals=True,
    ngals=10
).gen_sim()
```

```python
fig, axs  = plt.subplots(nrows=3, ncols=3, figsize=(12, 12))

for band_ind, band in enumerate(data):
    for epoch_ind, obs in enumerate(data[band]):
        ax = axs[band_ind, epoch_ind]
        ax.imshow(obs.image.array)
        ax.grid(False)
```
